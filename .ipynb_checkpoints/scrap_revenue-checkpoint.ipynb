```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np 
import re
import json
from pandas.io.json import json_normalize
```

```python
headers = {'User-Agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/78.0.3904.87 Safari/537.36'}
```

```python
url = 'https://en.wikipedia.org/wiki/Deloitte_Football_Money_League'
data = requests.get(url, headers = headers)
data.status_code
```

```python
soup = BeautifulSoup(data.content, "html.parser")
```

```python
wiki = soup.find_all('table', class_ = ["wikitable sortable"])
wiki17 = wiki[3]
wiki17
```

```python
lst_17 = []
for row in wiki17.find_all('tr'):
    for el in row.find_all('td'):
        lst_17.append(el.text)
lst_17_1 = lst_17[0::6]
lst_17_2 = lst_17[1::6]
lst_17_3 = lst_17[2::6]
lst_17_4 = []

lst_17_3 = [float(i)*1000000 for i in lst_17_3]

for i in range(30):
    lst_17_4.append("16/17")

columns = ["Placement", "TO", "Revenue", "SEASON"]

df_17 = pd.DataFrame([lst_17_1, lst_17_2, lst_17_3, lst_17_4]).T

df_17.columns = columns
df_17.head(3)
```
