```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np 
import re
import json
from pandas.io.json import json_normalize
```

```python
#rows = soup.find_all("table") #class_="table table--standings js-datatables dataTable no-footer")
#soup.find_all('table', class_="t1")
url17 = 'https://kassiesa.home.xs4all.nl/bert/uefa/data/method4/trank2017.html'
data17 = requests.get(url17).content
soup17 = BeautifulSoup(data17, "html.parser")


table17 = soup17.find_all('td', align="left")
text17 = [t.text for t in table17] 
text1617 = text17[1:101]
text1617
lst17 = []
position=[]
res_position = []

for i in range(len(text1617)):
    lst17.append("16/17")
    position.append(i+1)
    res_position.append(len(text1617)-i)
    
columns = ["Rating", "Res_Rating", "TO", "SEASON"]
df_17 = pd.DataFrame([position, res_position, text1617, lst17]).T
df_17.columns = columns
df_17
```
